```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = "gsarti/opus-mt-tc-en-pl"
```

```python
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
def translate(text):
    tokens = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    translated = model.generate(**tokens)
    return tokenizer.decode(translated[0], skip_special_tokens=True)
```

```python
print(translate("Job is fun"))
```

#Model wyszukiwanie Preview

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
offers = [
    "Praca zdalna jako programista Python na poziomie junior, mile widziane znajomość SQL i Django",
    "Praca na stanowisku Senior Java Developer w Katowicach. Wymagana znajomość Spring",
    "Programista C, oferujemy duża pensje i karnet na siłownie"
]
```

```python
embeddings = model.encode(offers, convert_to_tensor=True)
```

```python
input = "Szukam pracy jako Python Developer i chce karnet na siłownie"
input_embedding = model.encode(input, convert_to_tensor=True)
```

```python
cosine_scores = util.cos_sim(input_embedding, embeddings)[0]
```

```python
results = sorted(
    zip(offers, cosine_scores),
    key=lambda x: x[1],
    reverse=True
)
```

```python
for offer, score in results:
  print(f"({offer})  Ocena: {score}")
```
